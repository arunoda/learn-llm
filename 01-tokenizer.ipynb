## Tokenization 

LLMs only work with numbers. So, we need to convert text into numbers. This is the basic form of it.

```python
text = "hello world"
chars = sorted(list(set(text)))
chars
```

```python
# A map from char to integer
ctoi = {c: i for i, c in enumerate(chars)}
ctoi
```

```python
# A map from integer to the char
itoc = {i: c for c, i in ctoi.items()}
itoc
```

```python
encoded = [ctoi[c] for c in text]
encoded
```

```python
decoded_chars = [itoc[i] for i in encoded]
decoded = ''.join(decoded_chars)
decoded
```

## Final Notes

This is the most basic form of tokenizer. 

It only work with `text` only as the `char` only contains the chars contains in the `text`.
But the idea is the same.
